```python
import pandas as pd
import requests
import time
import json

# Load data
df = pd.read_csv(r'dataset/stream_data.csv')

anomaly_count = 0

for index, row in df.iterrows():
    
    # Prepare data
    payload = {
        'air_temp':      row['Air temperature [K]'],
        'process_temp':  row['Process temperature [K]'],
        'rpm':           row['Rotational speed [rpm]'],
        'torque':        row['Torque [Nm]'],
        'tool_wear':     row['Tool wear [min]']
    }
    
    # Send data to API
    response = requests.post(
        'http://127.0.0.1:5000/predict',
        json=payload
    )
    result = response.json()
    
    # Show anomalies for every 100 rows
    if result['anomaly']:
        anomaly_count += 1
        print(f" Row {index:>5} | anomaly | score: {result['score']:.4f} | RPM: {row['Rotational speed [rpm]']:.0f} | Torque: {row['Torque [Nm]']:.1f}")
    elif index % 100 == 0:
        print(f" Row {index:>5} | Normal  | score: {result['score']:.4f}")
    
    time.sleep(0.2)
    '''
    # Finish after 400 rows just for check 
    if index >= 400:
        print("=" * 50)
        print(f"Checked {index} rows | anomalies: {anomaly_count}")
        break 
    '''
```
